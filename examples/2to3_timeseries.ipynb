```python
## Convert timeseries file from py2 PySAR format to py3 PySAR format

import os
import sys
import h5py
import numpy as np
from pysar.utils import readfile, writefile
from pysar.objects import timeseries
from pysar.info import print_hdf5_structure

in_file = os.path.expanduser('~/insarlab/Galapagos/AlcedoEnvA2T061_PY2/timeseries_LODcor_ECMWF_demErr_refDate_plane.h5')
out_file = os.path.expanduser('~/insarlab/test/AlcedoEnvA2T061_timeseries.h5')

# read data
atr = readfile.read_attribute(in_file)
length, width = int(atr['LENGTH']), int(atr['WIDTH'])
f = h5py.File(in_file, 'r')
date_list = list(f['timeseries'].keys())
num_date = len(date_list)
ts_data = np.zeros((num_date, length, width), np.float32)
print('reading ...')
for i in range(num_date):
    date = date_list[i]
    ts_data[i, :, :] = f['timeseries/{}'.format(date)][:]
    sys.stdout.write('\r{}/{}'.format(i+1, num_date))
    sys.stdout.flush()
print('')

# prepare metadata
bperp = np.array([float(i) for i in atr['P_BASELINE_TIMESERIES'].split()], dtype=np.float32)
dates = np.array(date_list, np.string_)
for key in ['P_BASELINE_TIMESERIES', 'P_BASELINE_TOP_TIMESERIES', 'P_BASELINE_BOTTOM_TIMESERIES']:
    try:
        atr.pop(key)
    except:
        pass
atr['REF_DATE'] = date_list[0]

# write to HDF5 file
ts_obj = timeseries(out_file)
ts_obj.write2hdf5(data=ts_data, dates=dates, bperp=bperp, metadata=atr)
print('Done.')
```

```python
print_hdf5_structure(out_file)
```
