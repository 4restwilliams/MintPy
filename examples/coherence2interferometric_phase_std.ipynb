```python
##For distributed scatterers
## the marginal PDF of interferometric phase is given by (Tough et al., 1995)
## and Eq 4.2.32 (Hanssen, 2001)

import numpy as np
import matplotlib.pyplot as plt
import pysar.ifgram_inversion as ifginv

### PDF to Variance
plt.figure()
for L in [1,2,4,8,16]:
    var, coh = ifginv.phase_variance_ds(L)
    plt.plot(coh, np.sqrt(var)/np.pi*180, linewidth=2, label='L='+str(L))
plt.xlabel('Coherence')
plt.ylabel('Phase STD (deg)')
plt.xlim([0,1])
plt.ylim([0,120])
plt.legend()
plt.show()
```
