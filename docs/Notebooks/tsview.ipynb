## Plot displacement time-series in a notebook  

This notebook calls `tsview.py` in PySAR for interactive time-series visualization. The corresponding command line usage is print out for the reference. 

```python
%matplotlib notebook
import os
import matplotlib.pyplot as plt
from pysar.tsview import timeseriesViewer

# visualization functions
def tsview(fname, yx=None):
    """Plot input file using tsview.py script from PySAR"""
    cmd = 'tsview.py {} --ms 4 --noverbose'.format(fname)
    if yx is not None:
        cmd += ' --yx {} {}'.format(yx[0], yx[1])
    print(cmd)
    obj = timeseriesViewer(cmd)
    inps = obj.open()
    obj.figsize_img = [5, 4]
    obj.figsize_pts = [5, 2]
    obj.init_plot(inps)
    plt.show()
    return
```

```python
# use FernandinaSenDT128 as the example dataset
proj_dir = os.path.expanduser('~/work/FernandinaSenDT128/PYSAR')
os.chdir(proj_dir)
tsview('timeseries_ECMWF_ramp_demErr.h5', yx=(216, 310))
```
